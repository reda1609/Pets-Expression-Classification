# Transfer Learnin

From our code we can see that ResNet50 and MobileNetV1 tied for best accuracy. We will perform transfer learning on ResNet50

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader
import os
import time
import copy # For saving the best model

# --- Configuration ---
# TODO: Update these paths to match your Kaggle dataset structure
DATA_DIR = "/kaggle/input/pets-facial-expression-dataset/Master Folder" 
TRAIN_DIR = os.path.join(DATA_DIR, "train")
VAL_DIR = os.path.join(DATA_DIR, "valid") 
TEST_DIR = os.path.join(DATA_DIR, "test")

# Model parameters
NUM_CLASSES = 4 
BATCH_SIZE = 64
NUM_EPOCHS = 25 
LEARNING_RATE = 0.001

# Setup device
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {DEVICE}")
```

```python
# ImageNet statistics for normalization
imagenet_mean = [0.485, 0.456, 0.406]
imagenet_std = [0.229, 0.224, 0.225]

# Define transformations
# ResNet50 expects 224x224 images
data_transforms = {
    'train': transforms.Compose([
        transforms.RandomResizedCrop(224),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(imagenet_mean, imagenet_std)
    ]),
    'val': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(imagenet_mean, imagenet_std)
    ]),
    'test': transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(imagenet_mean, imagenet_std)
    ]),
}

# Create datasets
image_datasets = {}
if os.path.exists(TRAIN_DIR):
    image_datasets['train'] = datasets.ImageFolder(TRAIN_DIR, data_transforms['train'])
    NUM_CLASSES = len(image_datasets['train'].classes)
    print(f"Found {NUM_CLASSES} classes in the training set: {image_datasets['train'].classes}")
else:
    print(f"Training directory {TRAIN_DIR} not found. Cannot create training dataset.")
    # Exit or handle this error appropriately if training is essential

if os.path.exists(VAL_DIR):
    image_datasets['val'] = datasets.ImageFolder(VAL_DIR, data_transforms['val'])
else:
    print(f"Validation directory {VAL_DIR} not found. Validation phase will be skipped.")

if os.path.exists(TEST_DIR):
    image_datasets['test'] = datasets.ImageFolder(TEST_DIR, data_transforms['test'])
else:
    print(f"Test directory {TEST_DIR} not found. Cannot create test dataset.")


# Create dataloaders
dataloaders = {}
if 'train' in image_datasets:
    dataloaders['train'] = DataLoader(image_datasets['train'], batch_size=BATCH_SIZE, shuffle=True, num_workers=2) # Kaggle typically has 2 workers available
if 'val' in image_datasets:
    dataloaders['val'] = DataLoader(image_datasets['val'], batch_size=BATCH_SIZE, shuffle=False, num_workers=2)
if 'test' in image_datasets:
    dataloaders['test'] = DataLoader(image_datasets['test'], batch_size=BATCH_SIZE, shuffle=False, num_workers=2)

dataset_sizes = {x: len(image_datasets[x]) for x in image_datasets.keys()}
class_names = image_datasets['train'].classes if 'train' in image_datasets else []

if NUM_CLASSES == 0 and 'test' in image_datasets: # If train is missing but test exists for pure evaluation
    NUM_CLASSES = len(image_datasets['test'].classes)
    class_names = image_datasets['test'].classes
    print(f"Number of classes determined from test set: {NUM_CLASSES}, Classes: {class_names}")


if NUM_CLASSES == 0:
    print("Error: Could not determine the number of classes. Please check your dataset directories.")
else:
    print(f"Dataset sizes: {dataset_sizes}")
```

```python
# Load the pretrained ResNet50 model
model_resnet50 = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)

# Freeze all the parameters in the feature extractor
for param in model_resnet50.parameters():
    param.requires_grad = False

# Get the number of input features for the classifier
num_ftrs = model_resnet50.fc.in_features

# Replace the last fully connected layer with a new one for our number of classes
# The new layer's parameters will have requires_grad=True by default
model_resnet50.fc = nn.Linear(num_ftrs, NUM_CLASSES)

# Move the model to the device
model_resnet50 = model_resnet50.to(DEVICE)

print("ResNet50 model loaded and final layer replaced.")
# print(model_resnet50) # Uncomment to see model structure
criterion = nn.CrossEntropyLoss()
optimizer_ft = optim.Adam(model_resnet50.fc.parameters(), lr=LEARNING_RATE)
```

```python
def train_model(model, criterion, optimizer, num_epochs=10):
    since = time.time()
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    if 'train' not in dataloaders:
        print("Training dataloader not available. Skipping training.")
        return model

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase not in dataloaders: # Skip if val dataloader doesn't exist
                print(f"Skipping {phase} phase as dataloader is not available.")
                continue

            if phase == 'train':
                model.train()  # Set model to training mode
            else:
                model.eval()   # Set model to evaluate mode

            running_loss = 0.0
            running_corrects = 0

            # Iterate over data.
            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(DEVICE)
                labels = labels.to(DEVICE)

                # Zero the parameter gradients
                optimizer.zero_grad()

                # Forward
                # Track history if only in train
                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    # Backward + optimize only if in training phase
                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                # Statistics
                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            # Deep copy the model if it's the best validation accuracy so far
            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())
                print(f'New best val Acc: {best_acc:.4f}!')

        print()

    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    if 'val' in dataloaders: # Only print best_acc if validation was performed
        print(f'Best val Acc: {best_acc:4f}')
    else: # If no validation, the "best" model is the last one
        best_model_wts = copy.deepcopy(model.state_dict())


    # Load best model weights
    model.load_state_dict(best_model_wts)
    return model
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def evaluate_model(model, dataloader, criterion):
    if not dataloader:
        print("Test dataloader not available. Skipping evaluation.")
        return None, None, None

    model.eval()  # Set model to evaluation mode
    all_preds = []
    all_labels = []
    running_loss = 0.0
    running_corrects = 0

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(DEVICE)
            labels = labels.to(DEVICE)

            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            loss = criterion(outputs, labels)

            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)

            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    total_loss = running_loss / len(dataloader.dataset)
    total_acc = running_corrects.double() / len(dataloader.dataset)

    print(f'Test Loss: {total_loss:.4f} Acc: {total_acc:.4f}')
    return all_labels, all_preds, total_acc
```

```python
print("Starting model training...")
model_resnet50_ft = train_model(model_resnet50, criterion, optimizer_ft, num_epochs=NUM_EPOCHS)
```

```python
model_to_evaluate = model_resnet50_ft
```

```python
 # Use the model from training/loading

true_labels, predicted_labels, test_accuracy = evaluate_model(model_to_evaluate, dataloaders['test'], criterion)

if true_labels and predicted_labels:
    print("\nClassification Report:")
    print(classification_report(true_labels, predicted_labels, target_names=class_names, zero_division=0))

    print("\nConfusion Matrix:")
    cm = confusion_matrix(true_labels, predicted_labels)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()
```
